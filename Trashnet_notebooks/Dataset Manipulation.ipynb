# 5. Check how many files each class has and visualizing it

```python
import os
base_dir = r"C:\Users\Reah\Downloads\new2_trashnet"
hm = {}
total = 0
for sub_dir in os.listdir(base_dir):
    target_dir = base_dir + '/' + sub_dir
    files = os.listdir(target_dir)
    count = len(files)
    total += count
    hm[sub_dir] = count
print(hm)
```

```python
import matplotlib.pyplot as plt

plt.bar(range(len(hm)), list(hm.values()), align='center')
plt.xticks(range(len(hm)), list(hm.keys()))
plt.show()
```
